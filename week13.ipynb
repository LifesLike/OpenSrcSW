```python
# 실습 1

import csv
import matplotlib.pyplot as plt

with open('seoul.csv', 'r', encoding='cp949') as csv_file:
    data = csv.reader(csv_file)
    next(data)
    highest_temperature = []
    lowest_temperature = []

    for row in data:
        if row:
            date = row[0].split('-')
            if date[1] == '09' and date[2] == '14':
                highest_temperature.append(float(row[-1]))
                lowest_temperature.append(float(row[-2]))

    plt.plot(highest_temperature, color='pink')
    plt.plot(lowest_temperature, color='skyblue')
    plt.show()
```

```python
# 실습 2

import csv
import matplotlib.pyplot as plt

with open('seoul.csv', 'r', encoding='cp949') as csv_file:
    data = csv.reader(csv_file)
    next(data)
    january = []
    august = []

    for row in data:
        if row:
            date = row[0].split('-')
            if date[1] == '01':
                january.append(float(row[-1]))
            elif date[1] == '08':
                august.append(float(row[-1]))

    plt.hist(january, bins=100, color='b', label='Jan')
    plt.hist(august, bins=100, color='r', label='Aug')
    plt.legend()
    plt.show()
```
